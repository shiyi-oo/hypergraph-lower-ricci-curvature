# Two communities

## Balanced

```python
k=3
sizes = [15, 15]
p = [0.1, 0.001]
title = f"k={k} n={sizes} p={p}"
hyperedges, node_label, n_com_per_edge = generate_hsbm(k_uniform=k, block_sizes=sizes, community_edge_probs=p)
hlrc = compute_hlrc(hyperedges)
```

```python
sbm = pd.DataFrame()
sbm["hlrc"] = hlrc
sbm["n_c"] = n_com_per_edge
sbm["edge_label"] = ["inter" if n_c > 1 else "intra" for n_c in n_com_per_edge]

group1 = sbm.loc[sbm['edge_label'] == 'intra', 'hlrc']
group2 = sbm.loc[sbm['edge_label'] == 'inter', 'hlrc']

# Perform the Mann-Whitney U test
stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')
print("Statistic:", stat)
print("p-value:", p_value)
plot_edge_hlrc_boxplot(df=sbm,title=title,save=True)
```

```python
plot_graph(H=hyperedges, hlrc=hlrc, node_label=node_label,title=title, save=True)
```
