# Complete

```python
complete = generate_complete_hypergraph(n=5, k=3)
hlrc = compute_hlrc(complete)
np.unique(hlrc)
```

```python
H = xgi.Hypergraph(complete)
pos = xgi.pairwise_spring_layout(H,seed=123)
edge_color = cmap((hlrc[0]+1)/2)
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
xgi.draw(H, node_size=12,dyad_style="dashed", pos=pos,ax=ax,
         edge_fc=edge_color,
         edge_ec= "black",
         node_fc= "black",
         node_lw=0,
         alpha=0.4,
         hull = True,
         radius = 0.1,
         dyad_color=edge_color,
         rescale_sizes=False);
plt.savefig("figures/complete.pdf", bbox_inches='tight')
```

```python
tree = generate_hypertree(3, 2, 3)
hlrc = compute_hlrc(tree)
np.unique(hlrc)
```
